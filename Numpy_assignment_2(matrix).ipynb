```python
#THIS ONE GET ALL THE COEFFICIENTS OF JUST ONE PARTICULAR UNKNOWN BEFORE THE NEXT

import numpy as np

def get_coefficients(no_of_unknowns,no_of_unknowns_collected):
    '''Prompt the user to enter all the coefficients of an unknown(e.g x) in all the equation(s)'''
    x= []
    print('Enter all the coefficients of X%s: '%(no_of_unknowns_collected+1))
    for unknown in range(1,no_of_unknowns+1):
        k = int(input())
        x.append(k)
    x = np.array(x).reshape((no_of_unknowns,1))         
    return(x)

def get_constants(no_of_unknowns):
    '''Prompt the user to enter all the constants in all the equation(s)'''
    x= []
    print('Enter all the constants: ')
    for unknown in range(1,no_of_unknowns+1):
        k = int(input())
        x.append(k)
    x = np.array(x).reshape((no_of_unknowns,1))         
    return(x) 
    
    
#Get the number of unknowns to be solved from user
no_of_unknowns = int(input('Enter the number of unknowns to solve: '))
#using the mathematics logic of (x = A**-1 * B)
A = np.zeros((no_of_unknowns,no_of_unknowns))#creating an arbituary array for A
A = np.asmatrix(A)
B = np.zeros((no_of_unknowns,1))


no_of_unknowns_collected = 0
while no_of_unknowns_collected < no_of_unknowns:
    #iterating over to collect all the coefficeint of all unknowns
    A[:,(no_of_unknowns_collected)] = get_coefficients(no_of_unknowns,no_of_unknowns_collected)
    no_of_unknowns_collected +=1
print(A)

#collecting all the constants
print('Enter the constants values respectively: ')
B = get_constants(no_of_unknowns)
B = np.asmatrix(B)
print(B)

unknown_values = (A**(-1))*B
print('The values of the unknowns X(%s) are %s respectively'%(range(1,no_of_unknowns+1),unknown_values))
```
